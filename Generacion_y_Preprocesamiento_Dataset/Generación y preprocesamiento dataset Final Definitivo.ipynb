```python
import csv
from bs4 import BeautifulSoup
import urllib.request


def extraerJornadas_2019_2020():
    res = []
    url = "https://resultados.as.com/resultados/futbol/primera/2019_2020/calendario/" 
    fichero = urllib.request.urlopen(url)
    s = BeautifulSoup(fichero, "lxml")
    res += s.find_all("div", class_=["cont-modulo", "resultados"])
    return res




def cargar_datos_2019_2020():
    jornadas = extraerJornadas_2019_2020()
    lista_enlaces = []
    lista_jornadas = []
    lista_resultados = []
    lista_equipos_locales = []
    lista_equipos_visitantes = []
    for jornada in jornadas:
        jorn = jornada.find("h2").find("a").text
        lista_jornadas.append(jorn)
        enlacesPartidos = jornada.find("div", class_=["cont-resultados", "cf"]).find("table", class_="tabla-datos").find("tbody").find_all("tr")
        for enlace in enlacesPartidos:
            link = enlace.find("td", class_="col-resultado").find("a")["href"]
            resultado = enlace.find("td",class_="col-resultado finalizado").find("a").text
            equipo_local =  enlace.find("td", class_="col-equipo-local").find("a").find("span", class_="nombre-equipo").text
            equipo_visitante = enlace.find("td", class_="col-equipo-visitante").find("a").find("span", class_="nombre-equipo").text 
            lista_enlaces.append(link)
            lista_resultados.append(resultado)
            lista_equipos_locales.append(equipo_local)
            lista_equipos_visitantes.append(equipo_visitante)
    return lista_jornadas,lista_enlaces,lista_resultados,lista_equipos_locales,lista_equipos_visitantes  



def crear_csv_2019_2020():
    resultados= cargar_datos_2019_2020().__getitem__(2)
    equipoLocal=cargar_datos_2019_2020().__getitem__(3)
    equipoVisitante=cargar_datos_2019_2020().__getitem__(4)
    with open('dataset_Scraping_2019_2020.csv','wt')as csvfile: #Se puede sustituir para que se genere dónde haga falta/dataset3.csv','wt')as csvfile:   #Se puede sustituir para que se genere dónde haga falta
        writer = csv.writer(csvfile, delimiter=',', lineterminator='\n')
        writer.writerow(['Equipo_Local','Equipo_Visitante','Resultado'])  
        for eqLocal,eqVisit,result in zip(equipoLocal,equipoVisitante,resultados):
            writer.writerow([eqLocal,eqVisit,result])
            
extraerJornadas_2019_2020()
cargar_datos_2019_2020()
crear_csv_2019_2020()
```

### Métodos Auxiliares

```python
def obtener_Puntos_Locales(dataset):
    lista_puntos_locales=[]
    goles_locales = dataset['Goles_Local'].tolist()
    goles_visitantes = dataset['Goles_Visitante'].tolist()
    for gol_local,gol_vis in zip(goles_locales,goles_visitantes):
        if(gol_local > gol_vis):
            lista_puntos_locales.append(3)
        elif(gol_local < gol_vis):
            lista_puntos_locales.append(0)
        elif(gol_local == gol_vis):
            lista_puntos_locales.append(1)
    return lista_puntos_locales
```

```python
def obtener_Puntos_Visitantes(dataset):
    lista_puntos_visitantes=[]
    goles_locales = dataset['Goles_Local'].tolist()
    goles_visitantes = dataset['Goles_Visitante'].tolist()
    for gol_local,gol_vis in zip(goles_locales,goles_visitantes):
        if(gol_local > gol_vis):
            lista_puntos_visitantes.append(0)
        elif(gol_local < gol_vis):
            lista_puntos_visitantes.append(3)
        elif(gol_local == gol_vis):
            lista_puntos_visitantes.append(1)
    return lista_puntos_visitantes
```

```python
def obtener_resultado(dataset):
    resultados=[]
    goles_locales = dataset['Goles_Local'].tolist()
    goles_visitantes = dataset['Goles_Visitante'].tolist()
    for gol_local,gol_vis in zip(goles_locales,goles_visitantes):
        if(gol_local > gol_vis):
            resultados.append('1')
        elif(gol_local < gol_vis):
            resultados.append('2')
        elif(gol_local == gol_vis):
            resultados.append('x')
    return resultados
```

```python
dataset_definitivo_2020_2022['Puntos_Visitantes'] = obtener_Puntos_Visitantes(dataset_definitivo_2020_2022)
```

```python
dataset_definitivo_2020_2022['Resultado'] = obtener_resultado(dataset_definitivo_2020_2022)
```
